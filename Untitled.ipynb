```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
account=pd.read_csv("account.csv")
```

```python
account.head()
```

```python
account.info()
```

```python
account.frequency.value_counts()
```

```python
account.describe()
```

```python
sns.set_style("whitegrid");
sns.pairplot(account,x_vars=['account_id','district_id','date'],y_vars=['account_id','district_id','date'], hue="frequency", size=3);
```

```python
sns.violinplot(x='frequency',y='date',hue='frequency',data=account)
plt.title('Violin Plot')
plt.xlabel('date')
plt.ylabel('frequency')
plt.legend(loc='upper center')
plt.show();
```

```python
sns.relplot(data=account, x="account_id", y="district_id", hue="frequency")
plt.title('Replot Plot')
plt.xlabel('account_id')
plt.ylabel('district_id')
plt.show();
```

```python
sns.relplot(data=account, x="date", y="district_id",hue="frequency")
plt.title('Replot Plot')
plt.xlabel('account_id')
plt.ylabel('district_id')
plt.show();
```

```python
counts, bin_edges = np.histogram(account['date'], bins=10, density = True)
pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='PDF');
plt.plot(bin_edges[1:], cdf, label='CDF')
plt.title('CDF and PDF')
plt.xlabel('date')
plt.ylabel('Prob')
plt.legend()
plt.show();
```
