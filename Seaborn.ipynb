This function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. The kind parameter selects the underlying axes-level function to use:

scatterplot() (with kind="scatter"; the default)

lineplot() (with kind="line")

```python
#https://seaborn.pydata.org/generated/seaborn.relplot.html
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips,kind="scatter")
```

# Catplot()
This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. The kind parameter selects the underlying axes-level function to use:

### Categorical scatterplots:

stripplot() (with kind="strip"; the default)

swarmplot() (with kind="swarm")

### Categorical distribution plots:

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

### Categorical estimate plots:

pointplot() (with kind="point")

barplot() (with kind="bar")

countplot() (with kind="count")

```python
tips.head()
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker")
```

```python
# The strip plot is similar to a scatter plot. It is often used along with other kinds of plots for better analysis. 
# It is used to visualize the distribution of data points of the variable.
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker",kind="strip")
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker",kind="swarm")
```

# Categorical distribution plots:

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

### Like box plots, violin plots and boxen plots are used to represent comparison of a variable distribution (or sample distribution) across different "categories" (for example, temperature distribution compared between day and night, or distribution of car prices compared across different car makers).

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker",kind="box")
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker",kind="violin")
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue="smoker",kind="boxen")
```

```python
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', inner = None, data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips, ax = g.ax)
```

# Visualizing Distribution of the Data
distplot()

kdeplot()

jointplot()

rugplot()

```python
sns.distplot( a = tips.total_bill,hist=True,kde=False)
```

```python
sns.distplot( a = tips.total_bill,hist=False)
```

```python
#Plot marginal distributions by drawing ticks along the x and y axes.
#This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way.
sns.distplot( a = tips.total_bill,rug=True)
```

KDE plots have many advantages. Important features of the data are easy to discern (central tendency, bimodality, skew), and they afford easy comparisons between subsets. 

```python
sns.jointplot(x = 'size', y = 'total_bill', data = tips,kind="scatter")
```

```python
sns.jointplot(x = 'size', y = 'total_bill', data = tips,kind="hex")
```
